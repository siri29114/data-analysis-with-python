```python
#there are 5 subjects psycology,nonothecnology,medicine,orchology,anthroplogy two sudents have applied to do masters in any of these 5 fields with a line plot, show in which feild they can excel ,the cut off for each subject is 70
#sudent 1 = 67,45,73,71,57
#student2 = 43,67,68,53,85
#plotting two lists!
#syntax
##student1
#plt.plottype(x,y,label,color,markes,linewidth,linestye)
sub = ["psy","nano","med","onc","anthro"]
s1=[67,45,73,71,57]
s2=[43,67,68,53,85]
cut=[70,70,70,70,70]
#student1
plt.scatter(sub,s1,label="student1",color="k",marker='p') #line graph
#student2
plt.scatter(sub,s2,label="student2", color="y",marker='H')
#cutoff
plt.plot(sub,cut,label="cutoff",color="r",linestyle=":",linewidth=3)
plt.xlabel("subjects")
plt.ylabel("markes")
plt.legend()
```

```python
# when india vs australia match on live,lindi 268 as a target for australia  has given the runs for for every 5 overs has follows
#india 25,51,84,131,160,189,220,250,25,
#australia has completed 25 overs
#A=50,41,15,94,110,151
#plot.overs vs runs for both the countried with india has blue line and australia has a yellow line give appropriate labels and lagends,the marker for india is flag and australia is plus
#take overs from A range functions create runs as arrays
import numpy as np
india=np.array([25,51,84,131,160,189,220,250,267,297])
Australia=np.array([15,41,94,110,151])
ofi= np.arange(5,51,5)
ofa= np.arange(5,26,5)
plt.title("ind vs Aus")
#india
plt.plot(india,ofi,label="india",color="b",marker="<",linewidth=2) #line graph
#australia
plt.plot(Australia,ofa,label="Australia", color="y",marker="p")
plt.xlabel("overs")
plt.ylabel("runs")
plt.legend()
plt.grid()
```
