```python
#LC
#NEWLIST=[OUTPUT for if]
L1=[23,45,67,74,98]
res=[i*i for i in l1 if i >50]
print(res)
```

```python
#LC
#new list =[(st of if)if (cond) else (st of else)for]
l1=[23,45,54,64,65,6]
res=[i*i if i>50 else 'x' for i in l1]
print(res)
```
